```python
from statistics import mean, stdev
import math

out_deles = 'OUT081.csv'
out_nosso = 'OUT081_2.csv'

rel_deles = 'REL081.csv'
rel_nosso = 'REL081_2.csv'
```

```python
with open(out_deles, 'r') as f:
    data_out_deles = f.readlines()

with open(out_nosso, 'r') as f:
    data_out_nosso = f.readlines()
    
with open(rel_deles, 'r') as f:
    data_rel_deles = f.readlines()
    
with open(rel_nosso, 'r') as f:
    data_rel_nosso = f.readlines()
```

```python
out_certo = [list(map(float, l.split(','))) for l in data_out_deles]
out = [list(map(float, l.split(','))) for l in data_out_nosso]
```

```python
difs_x = []
difs_r = []
difs_f = []

for i in range(len(out_certo)):
    caso = out_certo[i][0]
    np = out_certo[i][1]
    nf = out_certo[i][2]
    
    dif_x = 100 * abs(out_certo[i][3] - out[i][3]) / out_certo[i][3]
    dif_r = 100 * abs(out_certo[i][4] - out[i][4]) / out_certo[i][4]

    try:
        dif_f = 100 * abs(out_certo[i][5] - out[i][5]) / out_certo[i][5]
    except ZeroDivisionError:
        dif_f = 100
        
    
    difs_x.append(dif_x)
    difs_r.append(dif_r)
    difs_f.append(dif_f)
    
    print(f"{caso:2.0f} | {np:3.0f} | {nf:3.0f} | {dif_x:7.1f} | {dif_r:4.1f} | {dif_f:6.1f}")
    
print(f"Erro X medio: ({mean(difs_x):7.2f} +- {stdev(difs_x):7.2f})%")
print(f"Erro R medio: ({mean(difs_r):7.2f} +- {stdev(difs_r):7.2f})%")
print(f"Erro F medio: ({mean(difs_f):7.2f} +- {stdev(difs_f):7.2f})%")
```
